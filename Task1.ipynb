```python
# create a new dataframe with only the columns required by the Waterloo Museum of Art
df = data[['Object ID','Department','Object Name','Title','Culture','Artist Nationality','Object Begin Date','Object End Date','Medium','Credit Line','Country']].copy()
```

```python
df.head()
```

```python
# checking the nulls values
df.isna().sum()
```

```python
df = df.drop_duplicates()
```

# Going into each column to check for anomalies

```python
len(df['Object ID'].unique())
```

```python
df[df['Department'].isnull()]
```

```python
# only 1 Na, filling with 'Unknown'
df['Department'] = df['Department'].fillna('Unknown')
```

```python
df[df['Object Name'].isnull()]
```

```python
data[data['Object Name'].isnull()]
```

```python
# using the Link Resource we can see that the Object name is 'grid with commercial adhesive tape'
# you could also fil, it with 'Unknown'
df.loc[165,'Object Name'] = 'Grid with commercial adhesive tape'
```

```python
df[df['Title'].isnull()]
```

```python
# since all the NaN title are Asian Art, I will fill it with 'Asian Art Unkown'
df['Title'] = df['Title'].fillna('Asian Art Unknown')
```

```python
df[df['Title'].isnull()]
```

```python
df[df['Culture'].isnull()]
```

```python
# over half of the dataset is unkown for the column 'Culture'
# going to use columns Region, Subregion and Country to fillna
df['Culture'] = df['Culture'].fillna(data['Region'])
df['Culture'] = df['Culture'].fillna(data['Subregion'])
df['Culture'] = df['Culture'].fillna(data['Country'])
```

```python
df[df['Culture'].isnull()]
```

```python
# still over 100 NaN values, will be fill with Unknown
df['Culture'] = df['Culture'].fillna('Unknown')

```

```python
df[df['Artist Nationality'].isnull()]
```

```python
# 115 NaN, we can use Country or Culture to help with the Nationality but it could result with incorrect data,
# I believe it is better to use 'Unknown' so that the information is not wrong.
df['Artist Nationality'] = df['Artist Nationality'].fillna('Unknown')
```

```python
df[df['Object Begin Date'].isnull()]
# no NaN
```

```python
df[df['Object End Date'].isnull()]
```

```python
#It appeards that the End Date is actually unknown for these 2 row
df['Object End Date'] = df['Object End Date'].fillna('Unknown')
```

```python
df[df['Medium'].isna()]
```

```python
#It appeards that the Medium is also unknown for these 2 row
df['Medium'] = df['Medium'].fillna('Unknown')
```

```python
df[df['Credit Line'].isna()]
# no NaN
```

```python
df[df['Country'].isna()]
```

```python
# first we will use the Culture column to determ the Country
df['Country'] = df['Country'].fillna(df['Culture'])
```

```python
# Now we will see the unique Country values and try to replace a few of them
df['Country'].unique()
```

```python
df['Country'] = df['Country'].replace({
    'American': 'United States', 'Japanese' : 'Japan', 'German' : 'Germany',
    'possibly Spanish' : 'possibly Spain','British' : 'Britain', 'Italian' : 'Italy',
    'French' : 'France', 'probably French' : 'probally France','Romanian' : 'Romenia',
    'European, Eastern' : 'Europe', 'European' : 'Europe', 'possibly Greek' : 'possibly Greece',
    'German, Silesia (Bunzlau)': 'Germany', 'Italian, Venice':'Italy', 'German, Hanau' : 'Germany',
    'probably German' : 'probably Germany', 'British, Chelsea': 'Britain', 'Spanish': 'Spanish',
    'Cypriot' : 'Cyprus', 'Italic': 'Unknown', 'Greek, Attic' : 'Greece','Minoan': 'Greece',
    'Minoan, Crete' : 'Greece','European or Middle Eastern' : 'Unknown', 'British, Stoke-on-Trent, Staffordshire' : 'Britain',
    'South Netherlandish' : 'Netherlands', 'present-day France' : 'France', 'India (Guler, Punjab Hills)' : 'India',
     'present-day Uzbekistan' : 'Uzbekistan', 'Southeastern Iran' : 'Iran' })
```
